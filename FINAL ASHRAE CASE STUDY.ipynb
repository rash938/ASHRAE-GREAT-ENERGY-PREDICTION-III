WE are plotting frequency vs meter type plot to see how many meters of each type are installed
across various sites and buildings.This will give us idea of numbers of meters installed across various sites and buildings.
the ylabel is frequency of each meter type and x label will be type of meter.

```python
df=pd.Series(train['meter'].value_counts())
index_=['electricity','chilledwater','steam','hotwater']
df.index=index_
plt.bar(index_,df)
plt.ylabel('frequency')
plt.title('frequency vs meter type')
```

1. electricity meters are of high number.
2. hotwater is least in number.
3. electricity is needed in all buildings all times so these type of meters are to be in high numbers. 
4. other meters depend on type of weather in a particular site so less in number.

next plot is log meter reading vs meter type plot. 
This plot gives average log meter reading for each type of meter.
this is a barplot.

```python
train['log_meter_reading'] = train.meter_reading.apply(np.log1p).astype('float32') #https://www.kaggle.com/pranavpateluk/energy-predictor-v3
grouped=train.groupby('meter')
l={0:'electricity',1:'chilledwater',2:'steam',3:'hotwater'}
data=list(l.values())
g=grouped['log_meter_reading'].mean()
plt.bar(data, g, align='center')
plt.ylabel('log meter reading')
plt.title('log meter reading vs meter type')
```

1. steam meter readings are highest followed by chilled water.
2. as we have seen in previous analysis that electricity meters are high in numbers but meter reading is low.
3. steam meter is most inefficient type meter.

now lets analyse meter_reading in each site_id by primary use.
To know the consumption of energy in each site_id we are plotting this graph.

```python

g=pd.DataFrame()
g['site_id']=train['site_id']
g['primary_use']=train['primary_use']
g['meter_reading']=train['meter_reading']
te5=g.groupby(['site_id', 'primary_use'])['meter_reading'].mean().unstack('primary_use')

ax=te5.plot(kind='bar', stacked=True,figsize=(14.5, 10.5))
ax.set_ylabel("mean meter_readings")
ax.set_title("stackedplot of meter reading by primary use per sit_id")
```

```python
'''
this function fills nan values of weather dataset.
 fill missing air temperature with mean temperature of day of the month. 
 Each month comes in a season and temperature varies lots in a season. So filling with yearly mean value is not a good idea.
 Data is missing for most of days and even many consecutive days. So, first, we calculate mean cloud_coverage of day of the month 
 then filled rest missing values with last valid observation.(fillna with the method='ffill').
 same is done for other features.
 '''
    


def fill_weather_dataset(df):
    
    
        
    # Reset Index for Fast Update
    df = df.set_index(['site_id','day','month'])

    air_temperature_filler = pd.DataFrame(df.groupby(['site_id','day','month'])['air_temperature'].mean(),columns=["air_temperature"])
    df.update(air_temperature_filler,overwrite=False)

    # Step 1
    cloud_coverage_filler = df.groupby(['site_id','day','month'])['cloud_coverage'].mean()
    # Step 2
    cloud_coverage_filler = pd.DataFrame(cloud_coverage_filler.fillna(method='ffill'),columns=["cloud_coverage"])

    df.update(cloud_coverage_filler,overwrite=False)

    due_temperature_filler = pd.DataFrame(df.groupby(['site_id','day','month'])['dew_temperature'].mean(),columns=["dew_temperature"])
    df.update(due_temperature_filler,overwrite=False)

    # Step 1
    sea_level_filler = df.groupby(['site_id','day','month'])['sea_level_pressure'].mean()
    # Step 2
    sea_level_filler = pd.DataFrame(sea_level_filler.fillna(method='ffill'),columns=['sea_level_pressure'])

    df.update(sea_level_filler,overwrite=False)

    wind_direction_filler =  pd.DataFrame(df.groupby(['site_id','day','month'])['wind_direction'].mean(),columns=['wind_direction'])
    df.update(wind_direction_filler,overwrite=False)

    wind_speed_filler =  pd.DataFrame(df.groupby(['site_id','day','month'])['wind_speed'].mean(),columns=['wind_speed'])
    df.update(wind_speed_filler,overwrite=False)

    # Step 1
    precip_depth_filler = df.groupby(['site_id','day','month'])['precip_depth_1_hr'].mean()
    # Step 2
    precip_depth_filler = pd.DataFrame(precip_depth_filler.fillna(method='ffill'),columns=['precip_depth_1_hr'])

    df.update(precip_depth_filler,overwrite=False)

    df = df.reset_index()
    
    return df
```

```python

'''
this function reduces memory usage by converting to datatypes which occupy less space in memory.
its a memory optimisation code.
helps the code to run faster and less memory space consumed.
'''


def reduce_mem_usage(df,verbose=True):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2    
    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)    
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose: print('Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction)'.format(end_mem, 100 * (start_mem - end_mem) / start_mem))
    return df
```
